### Review the problem

We were presented with the same formula in two forms.

$\color{red}{\lvert\overrightarrow{c}\rvert^2 =\lvert\overrightarrow{a}-\overrightarrow{b}\rvert^2 =}  \lvert\overrightarrow{a}\rvert^2 + \lvert\overrightarrow{b}\rvert^2 - 2\lvert\overrightarrow{a}\rvert\lvert\overrightarrow{b}\rvert\cos{\theta}$

and

$\color{red}{\lvert\overrightarrow{c}\rvert^2} =\lvert\overrightarrow{a}\rvert^2 + \lvert\overrightarrow{b}\rvert^2 - 2 \overrightarrow{a}\cdot{\overrightarrow{b}}$

So immediately, we can eliminate $\color{red}{\lvert\overrightarrow{c}\rvert^2}$.

$\color{blue}{\lvert\overrightarrow{a}\rvert^2 + \lvert\overrightarrow{b}\rvert^2} - 2\lvert\overrightarrow{a}\rvert\lvert\overrightarrow{b}\rvert\cos{\theta}$
$=\color{blue}{\lvert\overrightarrow{a}\rvert^2 + \lvert\overrightarrow{b}\rvert^2} - 2 \overrightarrow{a}\cdot{\overrightarrow{b}}$

Clearly $\color{blue}{\lvert\overrightarrow{a}\rvert^2 + \lvert\overrightarrow{b}\rvert^2}$ can go as well.

$\color{green}{-2}\lvert\overrightarrow{a}\rvert\lvert\overrightarrow{b}\rvert\cos{\theta}$
$= \color{green}{-2}\overrightarrow{a}\cdot{\overrightarrow{b}}$

Let's not be shy with the $\color{green}{-2}$ which leaves us with the following.

$\lvert\overrightarrow{a}\rvert\lvert\overrightarrow{b}\rvert\cos{\theta}$
$= \overrightarrow{a}\cdot{\overrightarrow{b}}$

Let's spell this out. It says that the length of vector $\overrightarrow{a}$ times the length of vector $\overrightarrow{b}$ times $\cos{\theta}$ is equal to the dot product of vectors $\overrightarrow{a}$ and $\overrightarrow{b}$.

If you're anything like me, this looks like an oversimplification that can't possibly be true. After all, where in the world did the trigonometric function $\cos{\theta}$ go. After all, the length of a vector is just pythagorean's theorem. There's no trig in that.

I think that if we can prove this, then we'll have a clearer understanding of what the dot product is and see how it relates to angles.

So to choke this down, we have to spell this out and prove it.

### Proof of the law of cosines in terms of vectors

Divide the triangle into two right triangles so the inserted segment is opposite of $\angle{\theta}$


For the side adjacent to $\angle{\theta}$, named $v$ in this proof, we can simply calculate the length of $v$.

$
\cos{\theta}=
  \frac{
    v
  }{
   \lvert\overrightarrow{b}\rvert
  }
$

$
\color{magenta}{
v =
\lvert\overrightarrow{b}\rvert\cos{\theta}
}
$

The remaining portion of vector $\overrightarrow{a}$ is as below

$
\color{blue}{
\lvert\overrightarrow{a}\rvert - v
}
$

We know from pythagorean's theorem the length of the side opposed of $\angle{\theta}$ (named $w$ here)

$
\lvert\overrightarrow{b}\rvert^2 =
 v^2
 +
 w^2
$

$\color{red}{w^2=\lvert\overrightarrow{b}\rvert^2-v^2}$

$
w =
\sqrt{
 \lvert\overrightarrow{b}\rvert^2 
 -
 v^2
}
$

We also know the length of vector $\overrightarrow{c}$ through pythagorean's theorem

$
\lvert\overrightarrow{c}\rvert^2 =
   (
     \color{blue}{
       \lvert\overrightarrow{a}\rvert - v
     }
   )^2
   +
   \color{red}{
     w^2
   }
$

Substitute for $\color{red}{w^2}$

$
\lvert\overrightarrow{c}\rvert^2 =
   \color{green}{
     (
       \lvert\overrightarrow{a}\rvert - v
     )^2
   }
   +
   \color{red}{
     \lvert\overrightarrow{b}\rvert^2 
     -
     v^2   
   }
$

Distribute (FOIL) the other side 

$
\lvert\overrightarrow{c}\rvert^2 =
   \color{green}{
     \lvert\overrightarrow{a}\rvert^2 
     -
     2\lvert\overrightarrow{a}\rvert{v}
     +
     v^2
   }
   +
   \lvert\overrightarrow{b}\rvert^2 
   -
   v^2   
$

Simplify and reorder to make it look more like the law of cosines

$
\lvert\overrightarrow{c}\rvert^2 =
   \lvert\overrightarrow{a}\rvert^2 
   +
   \lvert\overrightarrow{b}\rvert^2 
   -
   2\lvert\overrightarrow{a}\rvert\color{magenta}{v}
$

So we've proven the law of cosines in terms of vectors. This is very exciting.

$
\lvert\overrightarrow{c}\rvert^2 =
   \lvert\overrightarrow{a}\rvert^2 
   +
   \lvert\overrightarrow{b}\rvert^2 
   -
   2\lvert\overrightarrow{a}\rvert
   \color{magenta}{
     \lvert\overrightarrow{b}\rvert\cos{\theta}
   }
$


### Reminding ourselves of what our goal is!

In order to solve the problem we started with, we need to show that

$\lvert\overrightarrow{a}\rvert\lvert\overrightarrow{b}\rvert\cos{\theta}$
$= \overrightarrow{a}\cdot{\overrightarrow{b}}$

### Running the home stretch

To prove what we see above, we'll need to simply get rid of $\lvert\overrightarrow{c}\rvert$ as it just gets in the way.

As can be seen in the diagram, we know that

$
\overrightarrow{c}
=
\overrightarrow{a}
-
\overrightarrow{b}
$

and also that

$
\color{green}{
\lvert\overrightarrow{c}\rvert
=
\lvert
  \overrightarrow{a}
  -
  \overrightarrow{b}
\rvert
}
$

So the law of cosines in vector form can be rewritten as follows.

$
\color{green}{
  \lvert
    \overrightarrow{a}
    -
    \overrightarrow{b}
  \rvert
}
  ^2
=
   \color{blue}{
     \lvert\overrightarrow{a}\rvert
   }
   ^2 
   +
   \color{red}{
     \lvert\overrightarrow{b}\rvert
   }  
   ^2
   -
   2\lvert\overrightarrow{a}\rvert\lvert\overrightarrow{b}\rvert\cos{\theta}
$

We also know we can calculate the length of a vector by taking the square root of the sum of each component squared. In other word, pythagorean's theorem again.

$
\color{green}{
  \lvert
    \overrightarrow{a}
    -
    \overrightarrow{b}
  \rvert
}
  ^2
=
   \color{blue}{
     \sqrt{a_0^2+a_1^2}
   }
   ^2
   +
   \color{red}{
   \sqrt{b_0^2+b_1^2}
   }
   ^2
   -
   2
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$

Notice we didn't both subsituting all cases. This is because after I was almost done, I realized that I would only need to change them back. So it was just better to leave them alone here.

Now we eliminate the square root of the square

$
\color{green}{
  \lvert
    \overrightarrow{a}
    -
    \overrightarrow{b}
  \rvert
}
  ^2
=
   \color{blue}{
     a_0^2+a_1^2
   }
   +
   \color{red}{
     b_0^2+b_1^2
   }
   -
   2
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$

Now, let's find the length of what was $\color{green}{\overrightarrow{c}}$ but is now $\color{green}{
    \overrightarrow{a}
    -
    \overrightarrow{b}
}$

$
\color{green} {
  \sqrt{
    (a_0-b_0)^2
    +
    (a_1-b_1)^2
  }
}
^2
=
   a_0^2+a_1^2
   +
   b_0^2+b_1^2
   -
   2
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$

Then eliminate the redundant square of the power etc...

$
\color{green}{
(a_0-b_0)^2
+
(a_1-b_1)^2
}
=
   a_0^2+a_1^2
   +
   b_0^2+b_1^2
   -
   2
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$

Then, distribute (FOIL) out the polynomials

$
\color{red}{a_0^2}
-
2a_0b_0
+
\color{blue}{b_0^2}
+
\color{green}{a_1^2}
-
2a_1b_1
+
\color{brown}{b_1^2}
=
   \color{red}{a_0^2}
   +
   \color{green}{a_1^2}
   +
   \color{blue}{b_0^2}
   +
   \color{brown}{b_1^2}
   -
   2
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$

Simplify out $\color{red}{a_0^2}$, $\color{green}{a_1^2}$, $\color{blue}{b_0^2}$, and $\color{brown}{b_1^2}$.

$
\color{magenta}{
-
2
}
a_0b_0
\color{magenta}{
-
2
}
a_1b_1
=
   \color{magenta}{
   -
   2
   }
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$

Carve away the unnecessary $\color{magenta}{-2}$

$
\color{red}{a_0}
\color{blue}{b_0}
+
\color{red}{a_1}
\color{blue}{b_1}
=
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$

And now we should see the obvious pattern of the dot product we were looking for!!!

$
\color{red}{\overrightarrow{a}}\cdot{\color{blue}{\overrightarrow{b}}}
=
   \lvert\overrightarrow{a}\rvert
   \lvert\overrightarrow{b}\rvert
   \cos{\theta}
$
